```python
import tabula
import pandas as pd

# Specify the PDF file path
pdf_path = '/content/drive/MyDrive/Colab Notebooks/taxi data/data.pdf'

# Read the PDF file and extract tables as a list of DataFrames
tables = tabula.read_pdf(pdf_path, pages='all', multiple_tables=True)

# Process and work with the extracted tables
for idx, table in enumerate(tables, start=1):
    # Remove rows containing NaN values
    cleaned_table = table.dropna(axis=0, how='any')

    # Remove columns containing NaN values
    cleaned_table = table.dropna(axis=1, how='any')

    # Convert the cleaned table to a Pandas DataFrame
    df = pd.DataFrame(cleaned_table)

    # Check if the DataFrame and its columns have data
    if not df.empty and not df.columns.empty:
        # Print the DataFrame for the current table
        print(f"Table {idx}:\n{df}\n")
```

```python
import PyPDF2
import pandas as pd

# Step 1: Extract Data from PDF
pdf_file_path = '/content/drive/MyDrive/Colab Notebooks/taxi data/data.pdf'

with open(pdf_file_path, 'rb') as pdf_file:
    pdf_reader = PyPDF2.PdfReader(pdf_file)
    pdf_text = ''
    for page in pdf_reader.pages:
        pdf_text += page.extract_text()

# Step 2: Data Cleaning and Structuring
lines = pdf_text.split('\n')
header = lines[0].split()
lines = lines[1:]
data = []

for line in lines:
    values = line.split()
    if len(values) == len(header):
        data.append(dict(zip(header, values)))

# Step 3: Create a DataFrame and Save as CSV
df = pd.DataFrame(data)
df.to_csv('output_dataset.csv', index=False)

# Print the DataFrame
print(df)
```

```python
import PyPDF2

# Open the PDF file in read-binary mode
with open('/content/drive/MyDrive/Colab Notebooks/taxi data/data.pdf', 'rb') as pdf_file:
    # Create a PDF reader object using PdfReader
    pdf_reader = PyPDF2.PdfReader(pdf_file)

    # Get the number of pages in the PDF
    num_pages = len(pdf_reader.pages)

    # Loop through each page and extract text
    for page_num in range(num_pages):
        page = pdf_reader.pages[page_num]
        page_text = page.extract_text()

        # You can now work with the extracted text from each page
        print(f"Page {page_num + 1}:\n{page_text}\n")

# Close the PDF file
pdf_file.close()
```
